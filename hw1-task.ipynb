```python
def Level(x):
       Level = 'A'
       if x > 700000:
           Level = 'F'
       elif x > 400000:
           Level = 'E'
       elif x > 200000:
           Level = 'D'
       elif x > 100000:
           Level = 'C'
       elif x > 10000:
           Level = 'B'
       return Level    
```
